```python
df1.head(2)
```

```python
# Change datatype of budget
def clean_budget(budget):
  """
  Cleans the 'Budget' column by removing special characters
  and converting it to a float.
  """
  if isinstance(budget, str):  # Check if the value is a string
    budget = budget.replace('$', '').replace(',', '').split(' ')[0]  # Remove special characters
    try:
      return float(budget)  # Try converting to float
    except ValueError:
      return np.nan  # Handle cases with invalid budget format
  else:
    return budget  # Return original value if not a string

# Apply the clean_budget function to the 'Budget' column in df1
df1['Budget'] = df1['Budget'].apply(clean_budget)

# NaN is filling with avg of budget
df1['Budget'].fillna(df1['Budget'].mean(), inplace=True)
```
